```python
import glob
import re
from pathlib import Path
import yaml
from nltk.tokenize import RegexpTokenizer

def process_file(file_path):
    with open(Path(file_path)) as f:
        data = f.read()

    # Extract YAML metadata
    metadata = yaml.safe_load(re.findall(r"---\n(.*)\n---", data, re.DOTALL)[0])

    # Remove YAML metadata from data
    data = re.sub(r"---\n.*\n---", '', data, flags=re.DOTALL)

    # Tokenize the remaining text
    tokenizer = RegexpTokenizer(r'\w+') # \w+ matches any word character (equivalent to [a-zA-Z0-9_])
    tokens = tokenizer.tokenize(data)


    document = dict(metadata=metadata, tokens=tokens, path=file_path)

    return document

def process_files_in_directory(directory_path):
    # Get a list of all .txt files in the directory
    txt_files = glob.glob(f"{directory_path}/*.txt")

    # Process each file and store the results in a list
    processed_files = [process_file(file) for file in txt_files]

    return processed_files

processed_files = process_files_in_directory("data")

for file in processed_files:
    print("title: ",file['metadata']['title'], '\n tags:', file['metadata']['tags'], "\n length (tokens): ", len(file['tokens']))
```
